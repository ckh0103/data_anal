```python
import gdown
gdown.download("https://bit.ly/3736JW1", 'ns_book6.csv', quiet=False)
```

```python
# 통계처리의 정확성을 위해서 잘못된 데이터의 보정과 필요없는 데이터 제거 선처리 해야만 한다.

import pandas as pd
ns_book6 = pd.read_csv('ns_book6.csv', low_memory=False)
ns_book6.head()
```

```python
# 데이터프레임의 데이터들에 대해 정리해서 서술해 준다.

ns_book6.describe()
```

```python
sum(ns_book6['도서권수'] == 0)
```

```python
# 도서권수가 0이 아닌 데이터만 처리
ns_book7 = ns_book6[ns_book6['도서권수'] != 0]
```

```python
ns_book7.describe(percentiles=[0.3, 0.6, 0.9])
```

```python
# object - 실제적인 데이터
# freq : 빈도수
ns_book7.describe(include='object')
```

```python
# 평균 구하기

# x1, x2, x3
x = [10, 20, 30]

sum = 0

for i in range(3):
  sum += x[i]

print("평균", sum / len(x))
print(f"합계 {sum}")
```

```python
ns_book7.describe()
```

```python
# 평균 구하기

ns_book7['대출건수'].mean()
```

```python
# 중앙값 구하기

ns_book7['대출건수'].median()
```

```python
# 데이터의 개수가 홀수면 중앙의 값이 한개로 정해진다.
# 데이터의 개수가 짝수면 중앙에 위치한 2개의 숫자를 더해서 평균을 내준다.
temp_df = pd.DataFrame([1, 2, 5, 10])
temp_df.median()
```

```python
# 중복된 값을 제거하고 중앙값 구하기

ns_book7['대출건수'].drop_duplicates().median()
```

```python
print(ns_book7['대출건수'].min())
print(ns_book7['대출건수'].max())
```

```python
# 분위수 구하기 : 0(0%) ~ 1(100%)
# quantile() -> 괄호 안 아무것도 안쓸 시에는 50% 의 데이터를 구한다.
ns_book7['대출건수'].quantile(0.25)
```

```python
ns_book7['대출건수'].quantile([0.25, 0.5, 0.75])
```

```python
# 분산 구하기 - 평균에서 떨어진 정도의 숫자
# 10 * 10 -> 분산 / 10 -> 표준편차

#ns_book7['대출건수'].var()
print(ns_book7['대출건수'].mean())
print(ns_book7['대출건수'].std())
```

```python
# 최빈값 - 최대로 많이 나타나는 값

ns_book7['대출건수'].mode()
#ns_book7['도서명'].mode()
#ns_book7['발행년도'].mode()
```

```python
ns_book7.mean(numeric_only=True)
```
