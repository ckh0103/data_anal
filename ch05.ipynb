```python
import gdown
gdown.download("https://bit.ly/3pK7iuu", 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# figure : 도표의 크기 세팅
# figure(x축의크기, y축의 크기)
# 단위는 인치

plt.figure(figsize=(9,6))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 여러개의 그래프 그리기 - 서브 플롯
fig, axs = plt.subplots(2)

# 첫 번째 그래프의 설정
axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)

# 두 번째 그래프의 설정
axs[1].hist(ns_book7['대출건수'], bins=100)

# 작은 데이터 표시를 위해 로그처리
axs[1].set_yscale('log')

fig.show()
```

```python
# 여러개의 그래프 그리기 - 서브 플롯

fig, axs = plt.subplots(2, figsize = (6, 8))

# 첫번째 그래프의 설정
axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
# 첫번째 그래프 제목 설정
axs[0].set_title('scatter plot')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')


fig.show()
```

```python
# 여러개의 그래프 그리기 - 서브 플롯
# fig, axs = plt.subplots(2)
# fig, axs = plt.subplots(2, figsize=(6,8))
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# 첫번째 그래프 설정
axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
# 첫번째 그래프 제목 설정
axs[0].set_title("Scatter plot")
# x축 이름 설정 - 도서 권수
axs[0].set_xlabel("number of books")
# y축 이름 설정 - 대출건수
axs[0].set_ylabel("borrow count")

# 두번째 그래프 설정
axs[1].hist(ns_book7['대출건수'], bins = 100)
# 작은 데이터 표시를 위해 로그 처리
axs[1].set_yscale('log')
# 첫번째 그래프 제목 설정
axs[1].set_title("Histrogram")
# x축 이름 설정 - 대출 건수
axs[1].set_xlabel("borrow count")
# y축 이름 설정 - 빈도수 - 한글을 사용하려면 font 설치해서 사용해야한다.
axs[1].set_ylabel("frequency")

plt.show()
# fig.show()
```
