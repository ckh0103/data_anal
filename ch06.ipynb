```python
import matplotlib.pyplot as plt


# 해상도 설정
plt.rcParams['figure.dpi'] = 100
```

```python
# 샘플데이터
plt.plot([1, 4, 9, 16])

# 그래프 제목
plt.title('simple line graph')

plt.show()
```

```python
# 객체지향 API 방식 그래프 그리기
# plt 안에 plot을 꺼내면 subplot들로 처리된다.
fig, ax = plt.subplots()

ax.plot([1, 4, 9, 16])
ax.set_title('simple line graph')

fig.show()
```
